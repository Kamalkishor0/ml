## Exploratory Data Analysis

This section explores the relationship between different factors and road accidents using visualizations:

1.  Accidents by Year: Analyze the trend of accidents over time.
2.  Accidents by Severity: Examine the distribution of accident severity levels.
3.  Accidents by Weather Condition: Investigate the impact of weather conditions on accident occurrence.
4.  Accidents by Area (Rural or Urban): Compare accident rates in rural and urban areas.

```python
import matplotlib.pyplot as plt

# Count the number of accidents per year and other details
accidents_by_year = df['Year'].value_counts().sort_index()
accidents_by_severity = df['Accident_Severity'].value_counts().sort_index()
accidents_by_weather = df['Weather_Conditions'].value_counts().sort_index()
accidents_by_area = df['Urban_or_Rural_Area'].value_counts().sort_index()

# Create a 2x2 figure
fig, axes = plt.subplots(2, 2, figsize=(14, 10))  # 2 rows, 2 columns
fig.subplots_adjust(hspace=0.3, wspace=0.3)  # Add space between subplots

# **Draw horizontal and vertical dividers**
fig.add_artist(plt.Line2D([0.5, 0.5], [0, 1], color="black", linewidth=1))  # Vertical line
fig.add_artist(plt.Line2D([0, 1], [0.5, 0.5], color="black", linewidth=1))  # Horizontal line

# ---- Pie Chart (Accidents by Year) ----
accidents_by_year.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    ax=axes[0, 0]  # First subplot
)
axes[0, 0].set_ylabel('')  # Remove y-axis label
axes[0, 0].set_title("Accidents by Year", fontsize=12, fontweight='bold')

# ---- Bar Chart (Accidents by Severity) ----
accidents_by_severity.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    ax=axes[0, 1]  # Second subplot
)
axes[0, 1].set_ylabel('')
axes[0, 1].set_title("Accidents by Severity", fontsize=12, fontweight='bold')

# ---- Pie Chart (Accidents by Weather) ----
total_accidents = accidents_by_weather.sum()
threshold = 0.02
significant_weather = accidents_by_weather[accidents_by_weather / total_accidents > threshold]
removed_weather = accidents_by_weather[accidents_by_weather / total_accidents <= threshold]

# **Plot only significant weather conditions**
significant_weather.plot(
    kind='pie',
    autopct=lambda p: f'{p:.1f}%' if p > 2 else '',  # Hide <2% labels
    startangle=90,
    ax=axes[1, 0]  # Third subplot
)

axes[1, 0].set_ylabel('')
axes[1, 0].set_title("Accidents by Weather", fontsize=12, fontweight='bold')

legend_labels = [f"{w} ({count})" for w, count in removed_weather.items()]
axes[1, 0].legend(legend_labels, title="Other Weather Conditions (<2%)", loc="lower right", bbox_to_anchor=(1, -0.2))

# ---- Convert Histogram to Pie Chart (Accidents by Area) ----
accidents_by_area.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    ax=axes[1, 1]  # Fourth subplot
)
axes[1, 1].set_ylabel('')  # Remove y-axis label
axes[1, 1].set_title("Accidents by Area", fontsize=12, fontweight='bold')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
